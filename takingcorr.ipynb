```python
movies_df=pd.read_csv('./ml-latest-small/movies.csv')
ratings_df=pd.read_csv('./ml-latest-small/ratings.csv')
#ratings_df=ratings_df[:1000000]
```

```python
merged_df=ratings_df.merge(movies_df, on='movieId')
merged_df.head()
```

```python
ratings_avg=pd.DataFrame(merged_df.groupby('title')['rating'].mean())
ratings_avg.sort_values(by='rating',ascending=False).head()
```

```python
ratings_count=pd.DataFrame(merged_df.groupby('title')['rating'].count())
#ratings_count/=ratings_count.mean()
ratings_count.sort_values(by='rating',ascending=False)[:50]
```

```python
df=merged_df.pivot_table(merged_df,columns='title',values='rating')
df.head()
```

```python
from ast import literal_eval
features=['genres','keywords','production_companies','spoken_languages','cast','crew']
for feature in features:
    merged_df[feature]=merged_df[feature].apply(literal_eval)
```
