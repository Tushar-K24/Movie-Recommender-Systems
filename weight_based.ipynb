```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
Credits=pd.read_csv('tmdb_5000_credits.csv')
Movies=pd.read_csv('tmdb_5000_movies.csv')
```

```python
Credits.head()
```

```python
print(Movies.shape)
Movies.head()
```

```python
print(Credits.shape)
Credits=Credits.rename(columns={'movie_id': 'id'})
Credits.head()
```

```python
movies_merged=Credits.merge(Movies,on='id')
movies_merged.head()
```

```python
movies_merged.isnull().sum()
```

```python
movies_merged.columns
```

```python
movies_merged.drop(['title_x','title_y','homepage','tagline','status'],axis=1,inplace=True)
movies_merged.head()
```

```python
#weighted average: (rv+cm)/(v+m)

R=movies_merged['vote_average']
V=movies_merged['vote_count']
C=R.mean()
M=movies_merged['vote_count'].quantile(0.70)
movies_merged['weighted_average']=(R*V + C*M)/(V+M)
```

```python
movies_merged.head()
```

```python
movies_cleaned=movies_merged[['id','original_title','vote_average','vote_count','weighted_average','popularity']]
movies_ranked=movies_cleaned.sort_values('weighted_average',ascending=False)
movies_ranked[:15]
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='weighted_average',y='original_title',data=movies_ranked[:10])
plt.title('Movies rated on their Weighted Average')
plt.xlabel('Rating')
plt.ylabel('Movie Title')
```

```python
popularity=movies_cleaned.sort_values(by='popularity',ascending=False)
popularity[:10]
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='popularity',y='original_title',data=popularity[:10])
plt.title('Movies based on popularity')
plt.xlabel('Popularity')
plt.ylabel('Movie Title')
```

## Considering Popularity too

```python
#We will give equal weightage to weighted average and popularity
#For that, we will scale the parameters first
from sklearn.preprocessing import MinMaxScaler
```

```python
Scaler=MinMaxScaler()
```

```python
Scaled_ver=Scaler.fit_transform(movies_cleaned[['weighted_average','popularity']])
movies_normalized=pd.DataFrame(Scaled_ver,columns=['weighted_average','popularity'])
movies_normalized.head()
```

```python
movies_cleaned[['norm_weighted_average','norm_popularity']]=movies_normalized
movies_cleaned.head()
```

```python
Score=(movies_cleaned['norm_weighted_average']+movies_cleaned['norm_popularity'])/2
movies_cleaned['Score']=Score
movies_cleaned.head()
```

```python
movies_cleaned.sort_values(by='Score',ascending=False,inplace=True)
```

```python
movies_cleaned[:10]
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='weighted_average',y='original_title',data=movies_cleaned[:10])
plt.title('Weight based')
plt.xlabel('Weight')
plt.ylabel('Movies Title')
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='popularity',y='original_title',data=movies_cleaned[:10])
plt.title('Popularity based')
plt.xlabel('Popularity')
plt.ylabel('Movies Title')
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Score',y='original_title',data=movies_cleaned[:10])
plt.title('Score based')
plt.xlabel('Score')
plt.ylabel('Movies Title')
```
